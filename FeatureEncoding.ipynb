```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop='first',sparse_output=False)

X_train_ohe=ohe.fit_transform(X_train[['Sex','Embarked']]).astype('int64')
X_test_ohe=ohe.transform(X_test[['Sex','Embarked']]).astype('int64')
```

```python
X_train_ohe
```

```python
X_test_ohe
```

```python
train_nominal_encoded_df=pd.DataFrame(X_train_ohe,columns=ohe.get_feature_names_out(),
                                        index=X_train.index)
test_nominal_encoded_df=pd.DataFrame(X_test_ohe,columns=ohe.get_feature_names_out(),
                                        index=X_test.index)
train_nominal_encoded_df
```

```python
X_train=X_train.drop(columns=['Sex','Embarked']).join(train_nominal_encoded_df)
X_test=X_test.drop(columns=['Sex','Embarked']).join(train_nominal_encoded_df)
```
