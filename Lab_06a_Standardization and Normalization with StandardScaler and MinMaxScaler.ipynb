```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler

ss = StandardScaler()
mms = MinMaxScaler()
norm_dis_col = ['Age']
skewed_cols = ['Pclass', 'SibSp', 'Parch', 'Fare', 'Sex_male', 'Embarked_Q',
       'Embarked_S']

X_train[norm_dis_col] = ss.fit_transform(X_train[norm_dis_col])
X_test[norm_dis_col] = ss.transform(X_test[norm_dis_col])
X_train[skewed_cols] = mms.fit_transform(X_train[skewed_cols])
X_test[skewed_cols] = mms.transform(X_test[skewed_cols])
```

```python
X_train
```

```python
X_train.to_csv('X_train_scaled.csv',index=False)
X_test.to_csv('X_test_scaled.csv',index=False)
y_train.to_csv('y_train.csv',index=False)
y_test.to_csv('y_test.csv',index=False)
```
